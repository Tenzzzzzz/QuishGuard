```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
features = pd.read_csv("features_processed.csv")

plt.figure(figsize=(3, 3))

# The Fix: Added numeric_only=True to skip the URL strings
correlations = features.corr(numeric_only=True)['label'].sort_values(ascending=False)

# Drop the 'label' correlation with itself so the chart is cleaner
correlations = correlations.drop('label')

sns.barplot(x=correlations.values, y=correlations.index, palette='viridis')

plt.title("Feature Correlation with Malicious Label")
plt.xlabel("Correlation Coefficient")
plt.ylabel("Features")
plt.show()
```

```python
import csv

# 1. Read the raw text file
with open("dataset\\open_pish_data.txt", "r") as open_file:
    urls = [line.strip() for line in open_file.readlines() if line.strip()]

# 2. Write each URL to its own row in a single column
with open("dataset\\output_urls.csv", "w", newline='') as out_file:
    writer = csv.writer(out_file)

    # Optional: Add a header for the column
    writer.writerow(["URL"])

    for url in urls:
        # Putting 'url' inside brackets [url] keeps it in one column
        writer.writerow([url])

print("CSV created with one URL per row!")
```
