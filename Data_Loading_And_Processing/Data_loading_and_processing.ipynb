```python
import pandas as pd
import os
```

```python
first_data_file = pd.read_excel(r"..\dataset\pure_phishing_urls.xlsx")
```

```python
first_data_file["label"]=1
combined_df =first_data_file
print(combined_df.head())
print(combined_df["label"].value_counts())
```

```python
third_data_set=pd.read_csv(r"..\dataset\cleaned_phiusiil.csv")
```

```python

two_dataSets_combined_df = pd.concat([combined_df, third_data_set], ignore_index=True)
```

```python
print(two_dataSets_combined_df.tail())
print(two_dataSets_combined_df["label"].value_counts())
phishtank=pd.read_csv(r"../dataset/phishtank_urls.csv")
needed_col=phishtank[["url"]]
needed_col['label'] = 1
needed_col.info()
final_combined_with_new_data_df = pd.concat([two_dataSets_combined_df, needed_col], ignore_index=True)
final_combined_with_new_data_df = final_combined_with_new_data_df.sample(frac=1, random_state=42).reset_index(drop=True)
final_combined_with_new_data_df.describe()
```

```python
open_fpish_data=pd.read_csv(r"..\dataset\output_urls_from_open_phish_data.csv")
open_fpish_data = open_fpish_data.rename(columns={"URL":"url"})
open_fpish_data['label']=1
print(open_fpish_data.head())

all_data=pd.concat([open_fpish_data, final_combined_with_new_data_df], ignore_index=True)
all_data = all_data.sample(frac=1, random_state=42).reset_index(drop=True)
all_data
```

```python
redundant_count = all_data.duplicated(subset=['url']).sum()
print(f"Number of redundant URLs found: {redundant_count}")
```

```python
unique_urls_df = all_data.drop_duplicates(
    subset=['url'],
    keep='first')
```

```python
redundant_count = unique_urls_df.duplicated(subset=['url']).sum()
print(f"Number of redundant URLs found: {redundant_count}")
```

```python
unique_urls_df
```

```python
open_fpish_data2=pd.read_csv(r"..\dataset\output_urls_from_open_phish_2.csv")
open_fpish_data2['label']=1
open_fpish_data2.info()
all_datav2 = pd.concat([unique_urls_df, open_fpish_data2], ignore_index=True)
all_datav2.info()
```

```python
open_fpish_data3=pd.read_csv(r"..\dataset\output_urls_from_open_phish_2.csv")
open_fpish_data3['label']=1
open_fpish_data3.info()
final_dataset = pd.concat([all_datav2, open_fpish_data3], ignore_index=True)
final_dataset.info()
```

```python
unique_urls_df = final_dataset.drop_duplicates(
    subset=['url'],
    keep='first')
```

```python
unique_urls_df.to_csv(r'..\final_dataset\Final_clean_dataset.csv', index=False)
```
